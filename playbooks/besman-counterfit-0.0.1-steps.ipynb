### 1. Activate conda environment.

    $ conda activate counterfit

### 2. Move to the counterfit dir.

    $ cd counterfit

### 3. Activate CounterFit tool.
    
    $ counterfit

### 4. Check the available targets.
    
    $ list targets

### 5. Set target

    $ set_target <target_name>

### 6. Check all available attacks.

    $ list attacks

### 7. Set attack
    
    $ set_attack <attack_name>

### 8. Test the target at normal condition. Send a sample input and check the prediction.
    
    $ predict -i <sample_index>

    Every target has a sample input file in .npz format. The file has 10 sample inputs. User can choose sample index between 0-9. This command will send the input to the target and return the predict.

### 9. Now user need to set few parameters for the perform the attack.

    $ set_params <list_of_params>
  
    Execute ‘show options’ command to check the available parameters. Change the required parameters. For example:
    
        $ set_params --sample_index 5 --max_eval 5000 --max_iter 10
 
### 10. Now run the attack.
    
    $ run

### 11. Now again check the prediction to check if the attack is successful or not.
    
    $ predict -a

### 12. Save the report.
    
    $ save -r
